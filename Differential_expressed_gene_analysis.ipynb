## Importing libraries and reading data

Using pydeseq2 and sanbomics libraries for the analysis

```python
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats
from sanbomics.tools import id_map
from sanbomics.plots import volcano
from scipy.stats import mannwhitneyu

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
counts = pd.read_csv('./data/clean_rnaseq_counts.csv')
del counts["Unnamed: 0"]
counts = counts.set_index('Gene')
counts = counts.apply(pd.to_numeric, errors='coerce').astype(int)
counts
```

```python
# Filtering out genes where all values are 0
counts = counts[counts.sum(axis = 1) > 0]
counts
```

```python
# For analysis table needs to transposed
counts = counts.T
counts.head()
```

```python
# reading metadata
metadata = pd.read_excel('./data/clean_sample_sheet.xlsx')
del metadata["Unnamed: 0"]
metadata = metadata.set_index('SID')
metadata
```

## Performing DEG analysis

```python
dds = DeseqDataSet(counts=counts,
            metadata=metadata,
            design_factors="Condition")
```

```python
dds.deseq2()
```

```python
dds
```

```python
stat_res = DeseqStats(dds, contrast = ('Condition','AlcoholUseDisorder','Control'))
```

```python
stat_res.summary()
```

```python
# Getting stastistics for analysis
res = stat_res.results_df
res
```

```python
# For each gene assiging their symbols
mapper = id_map(species = 'human')
res['Symbol'] = res.index.map(mapper.mapper) 
res
```

```python
# Genes that are statistically significantly expressed
sigs = res[(res.padj < 0.05) & (abs(res.log2FoldChange) > 0.5)]
sigs
```

## Volcano plot

```python
volcano(res, symbol='Symbol')
```

In this plot it can be seen which genes are differentially expressed in which condition group. If gene is in top right field, it's differentially expressed in AUD (alcohol use disorder) patients and top left shows which genes are differentially expressed in control group.

Some things to note:
- There are much more differentially expressed genes for control patients than for AUD
- Top 5 genes for control group: AC004690.2, CNDP1, LRP2, HAGLR, AC009063.2
- Top 5 genes for AUD group: TGFBI, SERPINE1, CXCL3, CSF3, LIF

```python
DEG_counts = counts.loc[:, DEG_genes_ids]
DEG_counts.head()
```

```python
new_cols = []
for col in DEG_counts.columns:
    new_cols.append(renaming_dict[col])
DEG_counts.columns = new_cols
```

# Extra analysis: AUD or Control, which one has higher counts

```python
metadata = metadata.reset_index()
metadata.head()
```

```python
counts = counts.T
counts.head()
```

```python
control_values = []
AUD_values = []
for col in counts.columns:
    cond = list(metadata.loc[metadata["SID"] == col]["Condition"])[0]
    if cond == "Control":
        control_values.extend(list(counts[col]))
    elif cond == "AlcoholUseDisorder":
        AUD_values.extend(list(counts[col]))
print(len(control_values))
print(len(AUD_values))
```

```python
# Comparing values
print(np.mean(control_values))
print(np.mean(AUD_values))
```

```python
# Comparing whether values differ statistically
# Mann-Whitney U test (non-parametric alternative to t-test)
stat, p = mannwhitneyu(control_values, AUD_values, alternative='two-sided')
print(f"U-statistic: {stat}, p-value: {p}")
if p < 0.05:
    print("Samples have significantly different means.")
else:
    print("No significant difference between sample means.")
```
